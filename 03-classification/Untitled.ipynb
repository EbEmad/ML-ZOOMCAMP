### Feature importance

```python
female_mean = df_full_tain[df_full_tain.gender == 'female'].churn.mean()
print('gender == female:', round(female_mean, 3))

male_mean = df_full_tain[df_full_tain.gender == 'male'].churn.mean()
print('gender == male:  ', round(male_mean, 3))
```

```python
partner_yes = df_full_tain[df_full_tain.partner == 'yes'].churn.mean()
print('partner == yes:', round(partner_yes, 3))

partner_no = df_full_tain[df_full_tain.partner == 'no'].churn.mean()
print('partner == no :', round(partner_no, 3))
```

```python
female_mean / global_churn_rate
```

```python
male_mean / global_churn_rate
```

```python
df_group = df_full_tain.groupby(by='gender').churn.agg(['mean'])
df_group['diff'] = df_group['mean'] - global_churn_rate
df_group['risk'] = df_group['mean'] / global_churn_rate
df_group
```

```python
from IPython.display import display
global_mean = df_full_tain.churn.mean()
for col in cat_cols:
    df_group = df_full_tain.groupby(by=col).churn.agg(['mean'])
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['risk'] = df_group['mean'] / global_mean
    display(df_group)
```
